## ¿Qué es HTTP?

Significa Hypertext Transfer Protocol es el protocolo más popular ya que es usado en Internet para Comunicar paginas web que son elaboradas mediante documentos HTML.

Los documentos HTML contienen texto, que además, contienen enlaces, lo que hace posible que los documentos sean enlazados a diferentes páginas usando las ligas (links).

Por otra parte, los protocolos se refieren a un conjunto de reglas de comunicación que tienen que ser seguidos para que se realice la comunicación exitosa. Exiten diferentes formas en las que las computadoras se pueden comunicar y por lo tanto existen diversos protocolos.

Los protocolos más comunes son TCP/IP y además HTTP, FTP. HTTP permite comunicarse a través de internet mediante peticiones y a su vez obtiene respuestas de los recursos. HTTP es básicamente el fundamento de cualquier intercambio de datos en internet y además es un protocolo de tipo cliente-servidor.

Una petición iniciada por un cliente, lo que usualmente es un navegador web, y entonces obtinen de regreso una respuesta desde el servidor. 

HTTP es un protocolo sin estado (stateless), lo que significa es que, la petición actual no sabe nada sobre lo que se ha hecho en las peticiones anteriores.

## Métodos de petición HTTP (HTTP request methods)

Http fue diseñado para habilitar comunciación entre clientes y servidores, un servidor puede ser una computadora remota o la propia computadora del usuario cuando ejecuta software que brinda servicios o acceso a recursos. Un servidor, básicamente, es una computadora con grandes recursos que siempre está disponible (24 x 7).

HTTP trabaja como un protocolo de petición-respuesta entre un cliente y un servidor. El método que se usa para hacer una petición es el método GET y el servidor responde con un método RESPONSE. Se deben usar distintos métodos para obtener o modificar la información que se encuentra en el servidor.

Los métodos HTTP pueden ser vistos como verbos ya que indican la acción que se desea que realice el servidor sobre un determinado recurso.




| Método HTTP  | Descripción        |
|--------------|:------------------:|
| get          |  Se usa para pedir datos de un recurso específico |
| post         |  Se usa para enviar datos a un servidor |
| put          |  Se usa para actualizar datos existentes en un servidor |
| head         |  Parecido a  get, pero sin cuerpo de respuesta |
| delete       |  Elimina un recurso especificado del servidor |
| patch        |  Aplica modificaciones parciales a un servidor |
| Trace        |  Hace un eco de la petición recibida |




Las peticiones get pueden ser cacheadas, pueden ser marcadas como favoritas (bookmarked), nunca deben ser usadas con información sensible, tienen restricciones de longitud, y solo son usadas para obtener información.

El método post se usa para interactuar con los API de backend y enviarles información, ya que agrega nueva información al servidor. Algo importante a saber del método post es que este tipo de peticiones nunca son cacheadas, no se almacena en el historial del navegador, no pueden ser agregadas a favoritos (bookmarked), no tienen restricciones en la longitud de los datos.

el método put puede trabajar con las API para modificar datos existentes.

El método head es como el método request pero sin el cuerpo de la respuesta, es útil para para verificar lo que el método get regresará pero antes de hacer un método get, así que este método pide una respuesta que es identica a la respuesta de get pero sin el cuerpo de respuesta, lo que es útil para obtener la meta información que se escribe en los encabezados de la respuesta, pero sin tener que transportar el contenido entero.

El método trace es útil para que el cliente pueda ver cuáles cambios o adiciones han sido realizadas por los servidores.

A continuación se muestran algunas diferencias claves de los métodos más populares.



| Método GET   | Método POST        |
|--------------|:------------------:|
| no hace cambios          |  introduce nuevos datos |
| puede ser guardada en favoritos         |  No puede ser guardada en favoritos |
| puedes se cacheada          |  No puede ser cacheada |
| la petición permanece en el historial del navegador         |  la petición no permanece en el historial del navegador |
| la petición tiene restricción de longitud de datos       |  la petición no tiene restricción en la longitud de los datos |
| menos segura        |  mas segura |
| Los datos son visibles a todos desde la URL        |  Los datos no son visibles en la URL |



## Métodos seguros HTTP

Los métodos seguros HTTP son aquellos métodos que no causan ningún daño (modificación en el lado del servidor) debido a que su propósito principal es recuperar información y no cambiar el estado de la misma en el servidor. Como ejemplo tenemos los métodos Head; Get; Trace.

## Códigos de respuesta HTTP

Son códigos numericos que son regresados enseguida de una petición

Hay 5 niveles de códigos:

* 1 x (100,101)
* 2 x (200 - 206)
* 3 x (300 - 308)
* 4 x (400 - 451)
* 5 x (500 - 504)

Los códigos de respuesrta más comunes son:

| Mensaje  | Descripción        |
|--------------|:------------------:|
| 200 ok          |  Mensaje estanda para una petición exitosa de tipo HTTP |
| 201 created         |  Petición completada y nuevo recurso creado |
| 400 Bad request (404)          |  Sintáxis inválida (La petición no puede ser completada) |
| 403 Forbidden         |  El servidor se ha negado a responder a la petición |
| 405 Method not allowed       |  Método requerido no es soportado |
| 500 Internal Server Error        |  Servidor incapáz de completar la petición debido a algunos problemas |


Quienes usan estos códigos:

| Mensaje  | Descripción        |
|--------------|:------------------:|
| 200 ok          |  get, head, post, tracer |
| 201 created         |  post |
| 400 Bad request (404)          |  El seridor no pudo entender la petición (el servidor no pudo encontrar el recurso) |
| 403 Forbidden         |  El servidor entendió la petición pero se rehusa a autorizarla |
| 405 Method not allowed       |  Método requerido es conocido por el servidor pero ha sido deshabilitado |
| 500 Internal Server Error        |  El servidor encontró una condición inesperada que le impide completar la petición |
| 503 Service unavailable        |  El servidor no está listo para manejar la petición, el servidor esta sobrecargado o esta fuera de línea |
